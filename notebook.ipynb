```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting tampilan
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Load dataset utama
df1 = pd.read_csv('data/data1.csv')
df2 = pd.read_csv('data/data2.csv')
df3 = pd.read_csv('data/data3.csv')

# Tampilkan data
df1.head()
```

```python
# Info umum
print("Data 1:")
df1.info()
print("\nData 2:")
df2.info()
```

```python
df1.describe()
df1.isnull().sum()
cat_columns = df1.select_dtypes(include='object').columns

if len(cat_columns) > 0:
    for col in cat_columns:
        sns.countplot(data=df1, x=col)
        plt.title(f'Distribusi Kolom: {col}')
        plt.xticks(rotation=45)
        plt.show()
else:
    print("Tidak ada kolom kategorikal.")
```

```python
num_columns = df1.select_dtypes(include='number').columns

df1[num_columns].hist(bins=20, figsize=(15, 8), color='skyblue')
plt.suptitle("Distribusi Data Numerik", fontsize=16)
plt.show()
```

```python
corr = df1.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Matriks Korelasi")
plt.show()
```
